# Modéliser la propagation d’une maladie contagieuse

Motivé par l’actualité récente liée à la maladie COVID-19, ce projet a pour but l’étude de la propagation d’un virus au cours du temps au sein d’une population donnée. Modéliser cette propagation peut permettre d’aider à anticiper qualitativement et quantitativement la situation épidémiologique (déclenchement ou non d’une épidémie, nombre de personnes atteintes...). Le modèle peut également servir à étudier l’efficacité des différentes mesures envisagées pour ralentir la propagation de la maladie. On étudiera dans ce texte un modèle épidémiologique très classique : le modèle SIR. On commencera par présenter le modèle en détails et les hypothèses associées. La version discrète de ce modèle, qui est celle que nous allons étudier, consiste à construire des suites qui ont pour but de représenter la propagation de la maladie. Une fois le modèle établi, on étudiera les propriétés théoriques de ces suites. On pourra ensuite simuler numériquement ce modèle pour bien visualiser l’évolution et plus précisément voir comment les paramètres et conditions initiales influent sur la propagation de la maladie. Dans une seconde partie on introduira une quantité importante appelée taux de reproduction, qui a une interprétation assez simple. On verra pour commencer que la connaissance de cette valeur permet de déterminer si on a ou non épidémie, puis on verra que le nombre de personnes touchées par la maladie dépend du taux de reproduction. Enfin on s’intéressera à une interprétation de cette quantité.

## Le modèle SIR

### 1.1 Présentation du modèle
### Question 1 

**1. Que représentent S0,I0, R0 et M ?**



S0 représente le nombre d'individus suceptible d'ête infecté au temps t=0, I0 représente le nombre d'individu infecté au temps t=0 et R0 représente le nombre d'individus retirés au temps initial t=0, c'est à dire les personnes déjà guéries ou décédées.


M représente la population totale car S0 + I0 + R0 = M 

**2. 
(a) Que représente précisément α? 
On précisera en particulier son unité.** 


$\alpha$ apparaît dans la suite $I_{n+1}$ et $R_{n+1}$ : 

$$
I_{n+1} = I_n + \frac{q}{M}S_nI_n\Delta t - \alpha I_n \Delta t
$$

$$
R_{n+1} = R_n + \alpha I_n \Delta t
$$

On voit que le nombre d'infecté diminue de $\alpha I_n \Delta t$ et que le nombre de retiré diminue de $\alpha I_n \Delta t$. Cela signifie qu'entre deux instant une partie de la classe I, des infectés, qui la classe I pour aller à la classe R les retirés. 

Donc on pourrait assimiler alpha au taux de guérison ou taux de retrait des individus infectés. 

Pour respecter l'équilibre $\alpha \Delta t$ est dans la $[mesure]^{-1}$ de delta. On a alors [alpha] = $temps^{-1}$ .

**(b) À quoi correspondrait le cas α = 0?** 


Si $\alpha$ = 0 alors : $R_{n+1} = R_{n}$ . 

Donc personne ne quitte la classe des infectieux. Cela veut dire qu’aucun individu infecté ne guérit, ne meurt, ou ne cesse d’être contagieux.

**(c) Que se passerait-il dans ce cas (d’un point de vue mathématique et d’un point de vue épidémiologique)?**

Lorsque $\alpha$ = 0, d'un point de vue mathématique les trois systèmes deviennent : 

$$
S_{n+1} = S_n - \frac{q}{M} S_n I_n \Delta t
$$

$$
I_{n+1} = I_n + \frac{q}{M} S_n I_n \Delta t
$$

$$
R_{n+1} = R_n
$$

Les retirés restent constants, les suceptibles diminuent, et les infectés augmentent. Cela signifie que le modèle correspond à une maladie où les personnes infectées restent contagieuses indéfiniment.  Cela semble peu réaliste.

**3.
(a) Justifier l’hypothèse (H3).**


L'hypothèse (H3) suppose que la variation du nombre d'individus sains entre deux temps est proportionnelle au produit :

$$
S_n I_n
$$

Cette hypothèse est cohérente car une contamination a lieu lorsqu'un individu susceptible rencontre un individu infectieux. Plus il y a d'individus susceptibles ($S_n$), plus il y a de personnes pouvant être contaminées. Plus il y a d'individus infectieux ($I_n$), plus il y a de personnes pouvant transmettre la maladie.

Le nombre de nouvelles contaminations doit donc être proportionnel au nombre de rencontres possibles entre ces deux groupes, ce qui justifie :

$$
S_n I_n.
$$

**(b) Expliquer ce que représente précisément q. On précisera en particulier son unité.**

Le paramètre $q$ représente le taux de transmission de la maladie. Il mesure la capacité d'un individu infectieux à contaminer d'autres personnes.

Comme le terme

$$
\frac{q}{M}S_nI_n\Delta t
$$

représente le nombre de nouvelles contaminations pendant la durée $\Delta t$ dans $S_n$, l'unité de $q$ est :

$$
[q]=\frac{1}{\text{temps}}
$$
donc 
$$
[q]=\text{temps}^{-1}
$$
selon l'unité de temps utilisée.


**(c) À quoi correspondrait le cas q = 0?**

Si :

$$
q=0
$$

alors le terme de contamination disparaît :

$$
\frac{q}{M}S_nI_n\Delta t = 0.
$$

Le système deviendrait :

$$
\begin{cases}
S_{n+1}=S_n \\
I_{n+1}=I_n-\alpha I_n\Delta t \\
R_{n+1}=R_n+\alpha I_n\Delta t
\end{cases}
$$

La classe des suceptibles reste constante et la classe des infectés diminue. Aucune nouvelle contamination ne peut donc avoir lieu.



**(d) Que se passerait-il dans ce cas (d’un point de vue mathématique et d’un point de vue épidémiologique)?**

D'un point de vue mathématique, la suite $(S_n)$ reste constante, la suite $(I_n)$ décroît progressivement et la suite $(R_n)$ augmente progressivement.

D'un point de vue épidémiologique, les individus déjà infectés guérissent progressivement ou quittent la classe des infectieux, mais ils ne transmettent la maladie à personne. La maladie ou infection finit donc par disparaître et aucune épidémie ne peut se développer.

**4. Pensez-vous que l’hypothèse (H4) soit adaptée à toutes les maladies contagieuses? Est-elle adaptée pour le coronavirus?**

L'hypothèse (H4) suppose qu'un individu guéri ne peut plus être infecté. Cette hypothèse n'est pas adaptée à toutes les maladies contagieuses. Certaines maladies permettent des réinfections lorsque l'immunité disparaît ou lorsque le virus évolue.

Dans le cas du coronavirus, cette hypothèse est une approximation raisonnable sur une courte période mais elle n'est pas totalement réaliste sur le long terme car des réinfections sont possibles.



**5. ⋆Sur quel (ou quels) paramètre(s) du modèle les différentes mesures (confinement, port du masque, vaccination...) mises en place pendant l’épidémie de COVID-19 ont une influence?**

Les mesures sanitaires modifient principalement le paramètre q, car elles diminuent les contacts ou la probabilité de transmission. Le confinement réduit le nombre de contacts. Le port du masque réduit le risque de transmission. Les gestes barrières réduisent également la transmission.

La vaccination peut aussi modifier les conditions initiales en diminuant le nombre de personnes susceptibles: $ S_0. $

Elle peut également réduire indirectement la valeur de q en limitant la transmission de la maladie.

Les mesures sanitaires agissent donc principalement sur q et parfois sur les conditions initiales: $S_0,\quad I_0,\quad R_0.$

### Question 2

**1. Montrer que pour tout $n \in \mathbb{N}$ on a $S_n + I_n + R_n = M$. À quelle(s) hypothèse(s) du modèle ce résultat est-il lié ?**


On démontre ce résultat par récurrence sur $n$.

**Initialisation :** Pour $n=0$, on a par hypothèse $S_0 + I_0 + R_0 = M$.

**Hérédité :** Supposons que $S_n + I_n + R_n = M$ pour un certain $n$. On a alors :

$$
S_{n+1} + I_{n+1} + R_{n+1} = \left(S_n - \frac{q}{M}S_nI_n\Delta t\right) + \left(I_n + \frac{q}{M}S_nI_n\Delta t - \alpha I_n \Delta t\right) + \left(R_n + \alpha I_n \Delta t\right)
$$

Les termes $\frac{q}{M}S_nI_n\Delta t$ s'annulent et de même pour les termes $\alpha I_n \Delta t$. Il reste :

$$
S_{n+1} + I_{n+1} + R_{n+1} = S_n + I_n + R_n = M.
$$

Par principe de récurrence, $S_n + I_n + R_n = M$ pour tout $n \in \mathbb{N}$.

Ce résultat traduit le fait que la population totale ne change pas au cours du temps : un individu qui sort de la classe S rejoint la classe I, et un individu qui sort de la classe I rejoint la classe R, sans qu'aucun individu n'apparaisse ou ne disparaisse du système. C'est exactement ce que dit l'hypothèse **(H1)** : la population est isolée et on néglige les naissances et les décès non liés au virus.

**2. Montrer en utilisant la question ci-dessus, que pour tout $n \in \mathbb{N}$ on a $(S_n, I_n, R_n) \in [0, M]^3$. En quoi ce résultat mathématique est-il important pour le modèle ?**


On procède encore par récurrence, en supposant que $(S_n, I_n, R_n) \in [0,M]^3$.

**Pour $S_{n+1}$ :** On a $\Delta t < \frac{1}{q}$, donc $\frac{q}{M}I_n\Delta t \leq \frac{q}{M}\cdot M \cdot \Delta t = q\Delta t < 1$. Donc :

$$
S_{n+1} = S_n\left(1 - \frac{q}{M}I_n\Delta t\right)
$$

est le produit de deux termes positifs (puisque $1 - \frac{q}{M}I_n\Delta t > 0$), donc $S_{n+1} \geq 0$. De plus $S_{n+1} \leq S_n \leq M$.

**Pour $R_{n+1}$ :** On a $\Delta t < \frac{1}{\alpha}$, donc $\alpha \Delta t < 1$. Donc $R_{n+1} = R_n + \alpha I_n \Delta t \geq R_n \geq 0$, et $R_{n+1} \leq R_n + I_n \leq M$ .

**Pour $I_{n+1}$ :** Comme $S_{n+1} + I_{n+1} + R_{n+1} = M$ (question 1), on a $I_{n+1} = M - S_{n+1} - R_{n+1}$. Puisque $S_{n+1} \geq 0$ et $R_{n+1} \geq 0$ et $S_{n+1}+R_{n+1}\leq M$, on en déduit $I_{n+1} \in [0,M]$.

Donc $(S_{n+1}, I_{n+1}, R_{n+1}) \in [0,M]^3$, ce qui achève la récurrence.

Ce résultat garantit que le modèle reste **cohérent biologiquement** : à aucun instant le nombre d'individus dans une classe ne devient négatif ou ne dépasse la taille de la population totale. Sans cette propriété, les suites $S_n$, $I_n$, $R_n$ n'auraient plus de sens en tant qu'effectifs de population.

**3. La propriété mathématique précédente reste-t-elle vraie si $\Delta t > \min\left(\frac{1}{\alpha}, \frac{1}{q}\right)$ ?**

Non.
Si par exemple $\Delta t > \frac{1}{q}$, alors le facteur $\frac{q}{M}I_n\Delta t$ peut devenir strictement supérieur à 1 (si $I_n$ est suffisamment grand), et dans ce cas :

$$
S_{n+1} = S_n\left(1 - \frac{q}{M}I_n\Delta t\right) < 0,
$$

ce qui n'a aucun sens pour un effectif de population. De même si $\Delta t > \frac{1}{\alpha}$, le terme $\alpha I_n \Delta t$ peut dépasser $I_n$, ce qui rendrait $I_{n+1}$ négatif.

**4. Montrer que la suite $(S_n)_{n\in\mathbb{N}}$ est décroissante et que la suite $(R_n)_{n\in\mathbb{N}}$ est croissante. En déduire que les suites $(S_n)_{n\in\mathbb{N}}$, $(I_n)_{n\in\mathbb{N}}$ et $(R_n)_{n\in\mathbb{N}}$ sont convergentes. On notera $S_\infty, I_\infty, R_\infty$ leurs limites respectives.**

**Décroissance de $(S_n)$ :**

$$
S_{n+1} - S_n = -\frac{q}{M}S_nI_n\Delta t
$$

Comme $S_n \geq 0$, $I_n \geq 0$, $q > 0$ et $\Delta t > 0$ (question 2), on a $S_{n+1} - S_n \leq 0$. La suite $(S_n)$ est donc décroissante.

**Croissance de $(R_n)$ :**

$$
R_{n+1} - R_n = \alpha I_n \Delta t \geq 0
$$

car $\alpha > 0$, $I_n \geq 0$ et $\Delta t > 0$. La suite $(R_n)$ est donc croissante.

**Convergence :**

- $(S_n)$ est décroissante et minorée par $0$ (question 2), donc elle converge vers une limite $S_\infty \geq 0$.
- $(R_n)$ est croissante et majorée par $M$ (question 2), donc elle converge vers une limite $R_\infty \leq M$.
- Comme $S_n + I_n + R_n = M$ pour tout $n$ (question 1), on a $I_n = M - S_n - R_n$. Par opérations sur les limites, $(I_n)$ converge également, vers $I_\infty = M - S_\infty - R_\infty$.

Les trois suites sont donc convergentes.

**5. Montrer que nécessairement on a $I_\infty = 0$. Qu'est-ce que cela signifie en terme de propagation de la maladie ? Quelle valeur donne le nombre total de personnes qui auront été contaminées ?**

On part de la relation $S_{n+1} = S_n\left(1 - \frac{q}{M}I_n\Delta t\right)$, donc :

$$
\frac{q}{M}\Delta t \cdot I_n = 1 - \frac{S_{n+1}}{S_n} \quad \text{(si } S_n \neq 0\text{)}.
$$

On somme cette relation pour $n$ allant de $0$ à $N-1$. Comme $(S_n)$ est décroissante et positive, et que le terme $1 - \frac{S_{n+1}}{S_n}$ est positif et lié à la décroissance de $S_n$, on peut montrer (par comparaison avec une somme télescopique sur $\ln(S_n)$, ou simplement en utilisant que $S_{n+1}-S_n = -\frac{q}{M}S_nI_n\Delta t$ et que $(S_n)$ converge) que la série $\sum_{n\geq 0} I_n$ converge.

En effet, $(S_n)$ étant décroissante et convergente, la série télescopique $\sum (S_n - S_{n+1})$ converge (sa somme partielle vaut $S_0 - S_N \to S_0 - S_\infty$, qui est finie). Or $S_n - S_{n+1} = \frac{q}{M}S_nI_n\Delta t$. Si $S_\infty > 0$, alors $S_n$ ne tend pas vers 0 et reste minoré par une constante strictement positive à partir d'un certain rang, ce qui force $\sum I_n$ à converger (sinon $S_n - S_{n+1}$ ne tendrait pas vers 0 assez vite). Comme le terme général d'une série convergente tend vers 0 :

$$
I_n \longrightarrow 0, \quad \text{donc } I_\infty = 0.
$$

**Interprétation épidémiologique :** Le nombre d'infectieux finit toujours par s'annuler : la maladie s'éteint nécessairement à long terme, qu'il y ait eu épidémie ou non. Il n'existe pas de scénario où la maladie devient endémique et persiste indéfiniment dans ce modèle.

**Nombre total de personnes contaminées :** Comme $S_\infty + I_\infty + R_\infty = M$ et $I_\infty = 0$, on a $S_\infty + R_\infty = M$. Le nombre total de personnes qui auront été infectées à un moment ou un autre (en supposant $R_0=0$ au départ) est donc donné par $R_\infty$, c'est-à-dire :

$$
\text{Nombre total de personnes contaminées} = R_\infty = M - S_\infty.
$$

### Question 3

**1. On pose $X_n = \begin{pmatrix} S_n \\ I_n \\ R_n \end{pmatrix}$. Réécrire le système (1) sous forme vectorielle, c'est-à-dire donner une fonction $f : \mathbb{R}^3 \to \mathbb{R}^3$ telle que $X_{n+1} = f(X_n)$.**

On veut écrire $X_{n+1} = f(X_n)$ où $X_n = (S_n, I_n, R_n)$. D'après le système (1) :

$$
f\begin{pmatrix} S \\ I \\ R \end{pmatrix} =
\begin{pmatrix}
S - \dfrac{q}{M}SI\Delta t \\[6pt]
I + \dfrac{q}{M}SI\Delta t - \alpha I \Delta t \\[6pt]
R + \alpha I \Delta t
\end{pmatrix}
$$

On vérifie bien que la somme des trois composantes de $f(S,I,R)$ vaut $S+I+R$ (cohérent avec la question 2.1).

**2. Écrire un programme Python qui permet de visualiser sur un même graphique les évolutions des suites $(S_n)_{n\in[\![0,N]\!]}$, $(I_n)_{n\in[\![0,N]\!]}$ et $(R_n)_{n\in[\![0,N]\!]}$ en fonction de $n$.**

```python
import numpy as np
import matplotlib.pyplot as plt

def f(X, M, q, alpha, dt):
    """Fonction de transition du modèle SIR : X_{n+1} = f(X_n)."""
    S, I, R = X
    S_next = S - (q / M) * S * I * dt
    I_next = I + (q / M) * S * I * dt - alpha * I * dt
    R_next = R + alpha * I * dt
    return np.array([S_next, I_next, R_next])

def simulate_SIR(M, S0, I0, R0, q, alpha, N, T):
    """
    Simule le modèle SIR discret sur N pas de temps, entre t=0 et t=T.
    Renvoie les tableaux S, I, R ainsi que le pas de temps dt utilisé.
    """
    dt = T / N

    # On vérifie la condition de stabilité vue à la Question 2
    assert dt < min(1/q, 1/alpha), "Condition dt < min(1/q, 1/alpha) non respectée !"

    S = np.zeros(N + 1)
    I = np.zeros(N + 1)
    R = np.zeros(N + 1)
    S[0], I[0], R[0] = S0, I0, R0

    X = np.array([S0, I0, R0])
    for n in range(N):
        X = f(X, M, q, alpha, dt)
        S[n+1], I[n+1], R[n+1] = X

    return S, I, R, dt

def plot_SIR(S, I, R, dt, title="Modèle SIR"):
    """Affiche S_n, I_n, R_n en fonction de t_n = n*dt sur un même graphique."""
    N = len(S) - 1
    t = np.arange(N + 1) * dt

    plt.figure(figsize=(10, 5))
    plt.plot(t, S, label="$S_n$ (Susceptibles)", color="steelblue")
    plt.plot(t, I, label="$I_n$ (Infectieux)", color="orange", linestyle="--")
    plt.plot(t, R, label="$R_n$ (Retirés)", color="green", linestyle="--")
    plt.xlabel("$t_n$")
    plt.ylabel("populations")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

**3. Tester le programme avec les valeurs initiales $I_0 = 10$, $S_0 = 990$ et les paramètres $q = 2$, $\alpha = 0.8$, $N = 1000$ (le résultat obtenu est donné dans la figure 1 ci-dessous). Interpréter les résultats obtenus : la maladie s'est-elle beaucoup propagée ? Est-ce que toute la population a été contaminée ?...**

```python
M = 1000
S0, I0, R0 = 990, 10, 0
q, alpha = 2, 0.8
N = 1000
T = 20

S, I, R, dt = simulate_SIR(M, S0, I0, R0, q, alpha, N, T)
plot_SIR(S, I, R, dt, title="SIR : q=2, alpha=0.8")

print(f"dt = {dt}")
print(f"R0 (taux de reproduction) = q/alpha = {q/alpha:.2f}")
print(f"Nombre total de personnes contaminées (R_infini approché) ≈ {R[-1]:.0f}")
```

On obtient bien la même figure que la figure 1 de l'énoncé. On observe que la courbe de $S_n$ chute fortement (de 990 à environ 100), tandis que $R_n$ croît jusqu'à environ 900. La courbe $I_n$ présente un pic autour de $t \approx 4$, puis redescend vers 0, ce qui est cohérent avec le résultat $I_\infty = 0$ démontré à la question 2.5.

La maladie s'est donc beaucoup propagée : avec $\mathcal{R}_0 = q/\alpha = 2.5 > 1$, on observe bien une épidémie. Cependant, toute la population n'a pas été contaminée : il reste environ 100 personnes susceptibles qui n'ont jamais été infectées. L'épidémie s'arrête d'elle-même avant d'avoir touché toute la population, car le nombre de susceptibles devient trop faible pour entretenir la propagation (cohérent avec la question 4.3, où l'on montre que $I_n$ décroît dès que $S_n$ descend sous $M/\mathcal{R}_0$).

**4. Avec les mêmes conditions initiales, trouver des paramètres $\alpha \in ]0,1]$, $q \in ]0,M]$ tels que toute la population soit infectée pendant la période étudiée. Commenter la figure obtenue.**

Pour contaminer (quasiment) toute la population, il faut un taux de reproduction $\mathcal{R}_0 = q/\alpha$ très grand : une transmission forte ($q$ grand) combinée à une guérison lente ($\alpha$ petit). On augmente également $T$ pour laisser le temps à $I_n$ de redescendre vers $0$ (cf. Question 2.5).

```python
q2, alpha2 = 5, 0.05   # R0 = q/alpha = 100
N2 = 4000
T2 = 60

S2, I2, R2, dt2 = simulate_SIR(M, S0, I0, R0, q2, alpha2, N2, T2)
plot_SIR(S2, I2, R2, dt2, title="SIR : toute la population infectée (q=5, alpha=0.05)")

print(f"dt = {dt2}")
print(f"R0 (taux de reproduction) = {q2/alpha2:.1f}")
print(f"Nombre total de personnes contaminées (R_infini approché) ≈ {R2[-1]:.0f}")
print(f"S final ≈ {S2[-1]:.4f}")
```

Avec $q=5$ et $\alpha=0.05$ (soit $\mathcal{R}_0 = 100$), on observe sur la figure que $S_n$ chute extrêmement vite vers 0 : la quasi-totalité de la population susceptible est contaminée dès les premiers instants. Le pic de $I_n$ est beaucoup plus élevé et plus précoce que dans la simulation précédente.

Pour observer $R_n$ se stabiliser proche de $M$, il faut simuler sur une durée $T$ plus longue que dans les cas précédents : en effet $\mathcal{R}_0$ très grand accélère la phase de propagation, mais $\alpha$ très petit signifie que les infectieux restent contagieux longtemps, donc que la phase de décroissance de $I_n$ (vers $I_\infty = 0$, cf. Question 2.5) est elle aussi très lente. Avec $T=60$, on obtient $S_\text{final}\approx 0$ et $R_n$ proche de $M=1000$ : la quasi-totalité de la population finit par contracter la maladie.

Cela illustre que pour contaminer toute la population, il faut un taux de reproduction $\mathcal{R}_0$ très grand : une transmission très efficace ($q$ grand) combinée à une guérison très lente ($\alpha$ petit), donc une période d'infectiosité longue. Dans ce cas, presque chaque individu finit par croiser un infectieux avant l'extinction (lente) de l'épidémie.

**5. ⋆ Avec les mêmes conditions initiales, trouver des paramètres $\alpha \in ]0,1]$, $q \in ]0,M]$ tels que le nombre d'infectés décroisse au début de la simulation et commenter la figure obtenue.**

Si $\mathcal{R}_0 = q/\alpha \leq 1$ alors $(I_n)$ est décroissante dès le début. On choisit $q$ nettement plus petit que $\alpha$ pour avoir $\mathcal{R}_0$ bien inférieur à $1$.

```python
q3_, alpha3_ = 0.05, 1   # R0 = q/alpha = 0.05 << 1
N3 = 500
T3 = 10

S3, I3, R3, dt3 = simulate_SIR(M, S0, I0, R0, q3_, alpha3_, N3, T3)

# Graphique complet (S, I, R)
plot_SIR(S3, I3, R3, dt3, title="SIR : I décroissant dès le début (q=0.05, alpha=1)")

# Zoom sur I et R, car S_n (~990) écrase complètement I_n et R_n (~0-11) sur le même axe
t3 = np.arange(N3 + 1) * dt3
plt.figure(figsize=(10, 5))
plt.plot(t3, I3, label="$I_n$ (Infectieux)", color="orange")
plt.plot(t3, R3, label="$R_n$ (Retirés)", color="green", linestyle="--")
plt.xlabel("$t_n$")
plt.ylabel("populations")
plt.title("Zoom sur $I_n$ et $R_n$")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(f"dt = {dt3}")
print(f"R0 (taux de reproduction) = {q3_/alpha3_:.2f}")
print(f"Nombre total de personnes contaminées (R_infini approché) ≈ {R3[-1]:.0f}")
```

**6. ⋆ Par quoi peut-on approcher le nombre total de personnes qui auront contracté la maladie à partir des résultats de vos simulations numériques ?**

D'après la Question 2.5, on sait que $I_\infty = 0$ et que le nombre total de personnes ayant contracté la maladie est donné par $R_\infty = M - S_\infty$ (en supposant $R_0 = 0$ au départ).

Numériquement, on ne calcule jamais directement $R_\infty$ (qui est une limite quand $n \to +\infty$), mais on peut l'**approcher** par la dernière valeur de la suite simulée, $R_N$, à condition que $N$ (et donc $T = N\Delta t$) soit suffisamment grand pour que $I_N$ soit déjà très proche de $0$ (c'est-à-dire que la simulation ait eu le temps de converger).

$$
\text{Nombre total de personnes contaminées} \approx R_N \quad \text{(pour } N \text{ suffisamment grand)}.
$$

C'est exactement ce qu'on a fait dans les simulations précédentes en affichant `R[-1]` (ou `R2[-1]`, `R3[-1]`) : on utilise la dernière valeur calculée de $R_n$ comme approximation de $R_\infty$.

$$
I_{n+1} - I_n = \frac{q}{M}S_nI_n\Delta t - \alpha I_n\Delta t
$$

On factorise par $I_n\Delta t$ et on obtient :

$$
I_{n+1} - I_n = I_n\Delta t\left(\frac{q}{M}S_n - \alpha\right)
$$

On sort maintenant $\alpha$ en facteur dans la parenthèse :

$$
I_{n+1} - I_n = \alpha I_n\Delta t \left(\frac{q}{\alpha}\times\frac{S_n}{M} - 1\right)
$$

Or $\mathcal{R}_0 = \dfrac{q}{\alpha}$, donc :

$$
I_{n+1} - I_n = \alpha I_n\Delta t\left(\mathcal{R}_0\frac{S_n}{M} - 1\right)
$$

### Cas $\mathcal{R}_0 \leq 1$

On veut montrer que $(I_n)$ est décroissante, c'est-à-dire que :

$$
I_{n+1}-I_n \leq 0
$$

pour tout $n$.

On sait d'après les questions précédentes que :

$$
\alpha > 0, \qquad \Delta t > 0, \qquad I_n \geq 0
$$

donc :

$$
\alpha I_n \Delta t \geq 0.
$$

On sait aussi que $S_n \in [0,M]$, donc :

$$
\frac{S_n}{M} \leq 1.
$$

Comme $\mathcal{R}_0 \leq 1$ et $\dfrac{S_n}{M}\leq 1$, et que ces deux quantités sont positives, on peut multiplier les deux inégalités entre elles :

$$
\mathcal{R}_0 \times \frac{S_n}{M} \leq 1 \times 1 = 1.
$$

Donc :

$$
\left(\mathcal{R}_0\dfrac{S_n}{M} - 1\right) \leq 0.
$$

On a alors $I_{n+1} - I_n$ qui est le produit d'un terme positif ou nul ($\alpha I_n\Delta t$) et d'un terme négatif ou nul (la parenthèse), donc :

$$
I_{n+1} - I_n \leq 0.
$$

La suite $(I_n)$ est donc décroissante.

### Cas $\mathcal{R}_0 < 1$

On veut montrer que $(I_n)$ est strictement décroissante, c'est-à-dire :

$$
I_{n+1} - I_n < 0.
$$

Dans ce cas, $\mathcal{R}_0 < 1$ strictement et $\dfrac{S_n}{M}\leq 1$, donc :

$$
\mathcal{R}_0 \dfrac{S_n}{M} < 1.
$$

Ainsi :

$$
\left(\mathcal{R}_0\dfrac{S_n}{M}-1\right) < 0.
$$

Pour que le produit

$$
\alpha I_n \Delta t \left(\mathcal{R}_0\dfrac{S_n}{M}-1\right)
$$

soit strictement négatif, il faut également que :

$$
\alpha I_n \Delta t > 0.
$$

Comme $\alpha > 0$ et $\Delta t > 0$, cela revient à supposer :

$$
I_n > 0.
$$

Tant que $I_n > 0$, on a donc :

$$
I_{n+1} - I_n < 0,
$$

ce qui montre que la suite $(I_n)$ est strictement décroissante.
